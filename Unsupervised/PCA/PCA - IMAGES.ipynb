```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the image
image_path = 'PCA images.webp'
image = Image.open(image_path)
image_data = np.array(image)

# Flatten the image to apply PCA on the color channels individually
h, w, c = image_data.shape
flat_image = image_data.reshape(-1, c)

# Apply PCA to reduce to 2-3 principal components
pca = PCA(n_components=3)
pca_transformed = pca.fit_transform(flat_image)

# Reconstruct the image from the reduced PCA components
reconstructed_image = pca.inverse_transform(pca_transformed).reshape(h, w, c).astype(np.uint8)

# Display the original and reconstructed images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image_data)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(reconstructed_image)
axes[1].set_title("PCA-Reduced Image")
axes[1].axis("off")

plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the image
image_path = 'Animals.webp'
image = Image.open(image_path)
image_data = np.array(image)

# Flatten the image to apply PCA on the color channels individually
h, w, c = image_data.shape
flat_image = image_data.reshape(-1, c)

# Apply PCA to reduce to 2-3 principal components
pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(flat_image)


# Calculate reduction in data size
original_size = flat_image.size
reduced_size = pca_transformed.size
reduction_percentage = (1 - reduced_size / original_size) * 100

# Reconstruct the image from the reduced PCA components
reconstructed_image = pca.inverse_transform(pca_transformed).reshape(h, w, c).astype(np.uint8)

# Display the original and reconstructed images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image_data)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(reconstructed_image)
axes[1].set_title("PCA-Reduced Image")
axes[1].axis("off")

plt.show()


# Print the reduction information
print(f"Original Data Size: {original_size}")
print(f"Reduced Data Size: {reduced_size}")
print(f"Data Reduction: {reduction_percentage:.2f}%")
```
